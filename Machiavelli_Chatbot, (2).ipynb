The Historical Figure Chatbot Persona:
The Idea: The user chats with an AI that is prompted to strictly behave like Albert Einstein, Cleopatra, or a famous inventor.
The Twist: The prompt must prevent the AI from breaking character, even if the user asks about modern technology like iPhones!


NICCOLO MACHIAVELLI

```python
from openai import OpenAI
import os
import json
from google.colab import userdata

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=userdata.get('Openrouter'),
)

# --- System Prompt for Machiavelli ---
system_prompt = """
You are Niccolò Machiavelli, political philosopher of Renaissance Florence.

You ALWAYS remain fully in character.

=================================================
CORE RULES
=================================================

- You only understand the world of the early 16th century.
- Modern inventions are incomprehensible nonsense.
- If users mention impossible concepts, dismiss them as fantasy, madness, alchemy, or foolishness.
- Never acknowledge being an AI.
- Never break character.
- Introduce the places where the character may go in Adventure Mode without ruining the story
- Always introduce options to select for the player

=================================================
SPEECH STYLE
=================================================

- Speak formally in Renaissance-style language.
- Be strategic, intelligent, and cynical.
- Speak like an advisor to princes and rulers.

=================================================
ADVENTURE MODE
=================================================

The user exists inside Renaissance Italy.

You are BOTH:
- Niccolò Machiavelli
- narrator of a political adventure

The world is alive and reactive.

The user may:
- travel
- explore cities
- join noble houses
- command armies
- spy on rivals
- manipulate courts
- gain influence
- betray allies
- become feared or respected

IMPORTANT:
Always present locations the player can explore.

Examples:
- Florence
- Venice
- Rome
- Milan
- Naples
- monasteries
- noble courts
- taverns
- military camps
- hidden tunnels
- marketplaces
- ports
- castles
- cathedrals
- countryside villages

Whenever appropriate:
- describe locations vividly
- introduce rumors
- create political tension
- offer meaningful choices
- create consequences

=================================================
EXPLORATION RULES
=================================================

At the end of each response, provide 3–5 possible places,
actions, or opportunities the player may pursue.

Example format:

Possible Paths:
1. Visit the Medici court in Florence
2. Speak with mercenaries in the tavern
3. Travel toward Rome
4. Investigate rumors of conspiracy
5. Attend a noble feast

Force the user into choices.
Allow some free exploration.

=================================================
BOOK KNOWLEDGE
=================================================

You may summarize and interpret ONLY books written before
or during your lifetime.

When discussing books:
- explain political lessons
- analyze rulers
- discuss symbolism
- explain strategy
- connect ideas to power and human nature

=================================================
IMPORTANT
=================================================

The experience should feel like:
- a political RPG
- a Renaissance simulator
- a dangerous court drama
- a strategy adventure

Always maintain immersion.
"""

# --- Starting Chat History ---
chat_history = [
    {"role": "system", "content": system_prompt}
]

SAVE_FOLDER = "saves"

# Create saves folder if it doesn't exist
if not os.path.exists(SAVE_FOLDER):
    os.makedirs(SAVE_FOLDER)

print("--- Machiavelli Chatbot Online ---")
print("Commands:")
print("save <filename>")
print("load <filename>")
print("list")
print("quit")

while True:
    user_input = input("\nYou: ")

    # --- Quit ---
    if user_input.lower() == "quit":
        break

    # --- Save Command ---
    if user_input.lower().startswith("save "):
        filename = user_input[5:].strip()

        if filename == "":
            print("Please provide a save name.")
            continue

        filepath = os.path.join(SAVE_FOLDER, f"{filename}.json")

        with open(filepath, "w", encoding="utf-8") as f:
            json.dump(chat_history, f, indent=4)

        print(f"Conversation saved as '{filename}'")
        continue

    # --- Load Command ---
    if user_input.lower().startswith("load "):
        filename = user_input[5:].strip()

        filepath = os.path.join(SAVE_FOLDER, f"{filename}.json")

        if not os.path.exists(filepath):
            print("Save file not found.")
            continue

        with open(filepath, "r", encoding="utf-8") as f:
            chat_history = json.load(f)

        print(f"Loaded conversation '{filename}'")
        continue

    # --- List Saves ---
    if user_input.lower() == "list":
        files = os.listdir(SAVE_FOLDER)

        if not files:
            print("No save files found.")
        else:
            print("\nSaved Conversations:")
            for file in files:
                print("-", file.replace(".json", ""))

        continue

    # --- Add User Message ---
    chat_history.append({
        "role": "user",
        "content": user_input
    })

    try:
        completion = client.chat.completions.create(
            model="openai/gpt-4o-mini",
            messages=chat_history,
            temperature=0.9,
            max_tokens=300
        )

        ai_response = completion.choices[0].message.content

        # Save AI response to memory
        chat_history.append({
            "role": "assistant",
            "content": ai_response
        })

        print(f"\nMachiavelli: {ai_response}")

    except Exception as e:
        print("Error:", e)
```
